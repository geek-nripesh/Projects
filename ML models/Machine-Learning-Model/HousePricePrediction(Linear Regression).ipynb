2. Data preparation and cleaning

```python
df.drop(['date', 'street', 'city', 'statezip', 'country'], axis = 1, inplace = True)
df.head()
```

```python
x = df.columns[1:]
y = df.columns[0]
```

Checking presence of outliers using boxplot

```python
fig, axs = plt.subplots(12, figsize = (18,50))
j=0
for i in x:
    sns.boxplot(df[i],color='cyan', ax=axs[j])
    j+=1
```

```python
def countGraph(x):
    df[x].value_counts().plot(kind='bar')
    plt.title('number of '+x)
    plt.xlabel(x)
    plt.ylabel('Count')
```

```python
countGraph('bedrooms')
```

```python
countGraph('bathrooms')
```

```python
countGraph('floors')
```

Drawing scatterplot to check linearity of data

```python
def scPlot(x):
    fig, axs = plt.subplots(figsize = (25,10))
    plt.scatter(df['price'],df[x])
    plt.title("Price vs "+x)
    plt.show()
```

```python
for i in x:
    scPlot(i)
```

Removing columns with low correlation

```python
df.drop(['sqft_lot','condition','yr_built','yr_renovated'], axis = 1, inplace = True)
```

```python
x = df.columns[1:]
y = df.columns[0]
```

Splitting the data into training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(df[x], df[y], test_size = 0.20,random_state =14)
```

Creating the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

Model achieved the score of 54.8% which is the coefficient of determination

```python
lr.score(x_test,y_test)
```

Plotting graph for predicted and actual value of price againt other independent varaibles<br>
Orange : Predicted value<br>
Blue : Actual value

```python
predY = lr.predict(x_test)
```

```python
for i in x:
    fig, ax = plt.subplots()
    plt.scatter(x_test[i], y_test)
    plt.scatter(x_test[i], predY)
    plt.title("Price vs "+i)
    plt.show()
```
