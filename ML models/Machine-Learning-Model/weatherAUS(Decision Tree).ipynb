```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weatherAUS.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=["RainTomorrow"], inplace=True)
```

```python
df.shape
```

```python
df['year'] = pd.to_datetime(df['Date']).dt.year
```

```python
df.head()
```

```python
df.drop(['Date'], axis = 1, inplace = True)
df.head()
```

```python
train_data = df[df['year']<2015]
val_data = df[df['year']==2015]
test_data = df[df['year']>2015]
```

```python
print(train_data.shape)
print(val_data.shape)
print(test_data.shape)
```

```python
input_cols = list(train_data.columns)[0:-2]
target_cols = ['RainTomorrow']
```

```python
input_cols
```

```python
train_input = train_data[input_cols].copy()
val_input = val_data[input_cols].copy()
test_input = test_data[input_cols].copy()

train_target = train_data[target_cols].copy()
val_target = val_data[target_cols].copy()

test_target = test_data[target_cols].copy()
```

```python
# Numerical and categorical columns
num_col = train_input.select_dtypes(include=np.number).columns.tolist()
cat_col = train_input.select_dtypes('object').columns.tolist()
```

```python
print("numerical columns : ",num_col)
print("categorical columns : ",cat_col)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean').fit(df[num_col])
```

```python
train_input[num_col] = imputer.transform(train_input[num_col])
val_input[num_col] = imputer.transform(val_input[num_col])
test_input[num_col] = imputer.transform(test_input[num_col])
```

```python
train_input.head()
```

Scaling numeric features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sclr = MinMaxScaler().fit(df[num_col])
```

```python
train_input[num_col] = sclr.transform(train_input[num_col])
val_input[num_col] = sclr.transform(val_input[num_col])
test_input[num_col] = sclr.transform(test_input[num_col])
```

```python
train_input.head()
```

Data Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown="ignore").fit(df[cat_col])
```

```python
encoder_col = list(encoder.get_feature_names(cat_col))
```

```python
encoder_col
```

```python
train_input[encoder_col] = encoder.transform(train_input[cat_col])
val_input[encoder_col] = encoder.transform(val_input[cat_col])
test_input[encoder_col] = encoder.transform(test_input[cat_col])
```

```python
xTrain = train_input[num_col+encoder_col]
xTest = test_input[num_col+encoder_col]
xVal = val_input[num_col+encoder_col]
```

```python
xTrain
```

```python
xTrain.isna().sum()
```

```python
train_target.isna().sum()
```

Training the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(random_state=0)
```

```python
dtc.fit(xTrain,train_target)
```

```python
train_pred = dtc.predict(xTrain)
train_pred
```

Evaluating the model

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(train_pred, train_target)
```

```python
train_Prob = dtc.predict_proba(xTrain)
train_Prob
```

```python
dtc.score(xVal, val_target)
```

Visualization
